## CPU Computer Hardware

* CPU Computer Hardware; excludeti din dataset coloanele: vendor name, model name, estimated relative performance; se va estima coloana "published relative performance".

```python
CPU_data: pd.DataFrame = pd.read_csv("data/machine.data", delimiter=",", header=None)
CPU_data = CPU_data.drop([0, 1, 9], axis = 1)
#CPU_data = CPU_data.drop(1, axis = 1)
#CPU_data = CPU_data.drop(9, axis = 1)
CPU_data.head()
```

```python
CPU_X: np.ndarray = CPU_data.iloc[:, :-1].values
CPU_y: np.ndarray = CPU_data.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression, LassoLars, ARDRegression, PassiveAggressiveRegressor, TheilSenRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, DotProduct
from sklearn.svm import SVR

from sklearn.model_selection import cross_validate, GridSearchCV, RandomizedSearchCV

from typing import List, Dict, Union, Any
```

```python
RegressionModel = Union[LinearRegression, LogisticRegression, LassoLars, ARDRegression, PassiveAggressiveRegressor, TheilSenRegressor,
                       KernelRidge, GaussianProcessRegressor, SVR]
SearchModel = Union[GridSearchCV, RandomizedSearchCV]
```

```python
def populate_dataFrame(dataFrame: pd.DataFrame, scores: Dict[str, List[float]], regression_model: str, grid_search: str, index: int) -> None:
    '''
    Function to populate a Pandas Dataframe given as parameter depending on a scores dictionary.
    :param dataFrame: pandas Dataframe to be populated
    :param scores: dictionary with keys of type string and values of type float that contains the values to be put in :param dataFrame:
    :param regression_model: string representing the name of the regression model used for finding the scores
    :param grid_search: string representing the name used for searching the best hyperparameters
    :param index: integer representing the line of :param dataFrame: that will be populated
    :returns None:
    '''
    dataFrame.iloc[index, 0] = regression_model
    dataFrame.iloc[index, 1] = grid_search
    dataFrame.iloc[index, 2] = scores['test_neg_mean_absolute_error'].mean()
    dataFrame.iloc[index, 3] = scores['test_neg_mean_squared_error'].mean()
    dataFrame.iloc[index, 4] = scores['test_neg_median_absolute_error'].mean()
    dataFrame.iloc[index, 5] = scores['train_neg_mean_absolute_error'].mean()
    dataFrame.iloc[index, 6] = scores['train_neg_mean_squared_error'].mean()
    dataFrame.iloc[index, 7] = scores['train_neg_median_absolute_error'].mean()
    dataFrame.iloc[index, 8] = scores['fit_time'].mean()
    dataFrame.iloc[index, 9] = scores['score_time'].mean()
```

### LinearRegression

```python
from sklearn import metrics
```

```python
regression: RegressionModel = LinearRegression()
parameter_grid: Dict[str, List[Any]] = {"fit_intercept":[True, False], "normalize":[True, False], "n_jobs":[None, 2, 4, 6]}
```

```python
grid_search: SearchModel = GridSearchCV(estimator=regression, param_grid=parameter_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(CPU_X, CPU_y)
best_parameters: Dict[str, List[Any]] = grid_search.best_params_
print(best_parameters)
```

```python
tunned_regression: RegressionModel = LinearRegression(**best_parameters)
grid_scores: Dict[str, List[float]] = cross_validate(tunned_regression, CPU_X, CPU_y, cv=5, scoring=['neg_mean_absolute_error', 'neg_mean_squared_error', 'neg_median_absolute_error'], return_train_score=True)
populate_dataFrame(CPU_model_df, grid_scores, 'LinearRegression', 'GridSearchCV', 0)
```

```python
random_search: SearchModel = RandomizedSearchCV(estimator=regression, param_distributions=parameter_grid, cv=5, scoring='neg_mean_squared_error')
random_search.fit(CPU_X, CPU_y)
best_parameters: Dict[str, List[Any]] = random_search.best_params_
print(best_parameters)
```

```python
tunned_regression: RegressionModel = LinearRegression(**best_parameters)
random_scores: Dict[str, List[float]] = cross_validate(tunned_regression, CPU_X, CPU_y, cv=5, scoring=['neg_mean_absolute_error', 'neg_mean_squared_error', 'neg_median_absolute_error'], return_train_score=True)
populate_dataFrame(CPU_model_df, random_scores, 'LinearRegression', 'RandomizedSearchCV', 1)
CPU_model_df.head(n=10)
```

&emsp;&emsp;Regresia caut relaii ntre variabile.

&emsp;&emsp;De exemplu, putei observa civa angajai ai unei companii i putei ncerca s nelegei modul n care salariile lor depind de anumite caracteristici, cum ar fi experiena, nivelul de educaie, rolul, oraul n care lucreaz i aa mai departe.

&emsp;&emsp;n mod similar, putei ncerca s stabilii o dependen matematic a preurilor caselor in functie de suprafeele lor, numrul de dormitoare, distanele de centrul oraului i aa mai departe.

&emsp;&emsp;n general, n regresie, luai n considerare de obicei un fenomen de interes i avei o serie de observaii. Fiecare observaie are dou sau mai multe caracteristici. Urmnd presupunerea c (cel puin) una dintre caracteristici depinde de celelalte, ncercai s stabilii o relaie ntre ele.

&emsp;&emsp;Cu alte cuvinte, trebuie s gsii o funcie care s asocieze suficient de bine unele caracteristici sau variabile la altele.

&emsp;&emsp;Caracteristicile dependente se numesc variabile dependente, ieiri sau rspunsuri.

&emsp;&emsp;Caracteristicile independente sunt numite variabile independente, intrri sau predictori.

&emsp;&emsp;Problemele de regresie au de obicei o variabil dependent continu i nelimitat. Cu toate acestea, intrrile pot fi date continue, discrete sau chiar categorice, cum ar fi sexul, naionalitatea, marca etc.

&emsp;&emsp;Regresia este unul dintre cele mai importante domenii n statistica i machine learning. Exist multe metode de regresie disponibile. ***Regresia liniar*** este una dintre ele.

&emsp;&emsp;***Regresia liniar*** este probabil una dintre cele mai importante i utilizate pe scar larg tehnici de regresie. Este printre cele mai simple metode de regresie. Unul dintre principalele sale avantaje este uurina interpretrii rezultatelor.

## Decision Tree Regression

***Link catre documentatia oficiala***: [https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html)

&emsp;&emsp;Decision Tree este un instrument de luare a deciziilor care folosete o structur de arbore asemntoare unei diagrame sau este un model de decizii i al tuturor rezultatelor posibile ale acestora, inclusiv rezultatele, costurile de intrare i utilitatea.

&emsp;&emsp;Algoritmul Decision Tree se ncadreaz n categoria algoritmilor de nvare supervizata. Funcioneaz att pentru variabilele de ieire continue, ct i pentru cele categorice.

&emsp;&emsp;Ramurile/muchiile reprezint rezultatul nodului, iar nodurile au fie:

&emsp;&emsp;&emsp;&emsp;1.Conditii (Noduri de decizie)

&emsp;&emsp;&emsp;&emsp;2.Rezultate (Noduri frunza)

&emsp;&emsp;Ramurile/marginile reprezint adevrul/falsitatea declaraiei i un exemplu de luare a deciziei este in poza de mai jos, care arat un arbore de decizie care evalueaz cel mai mic dintre trei numere:

<img src="./images/decision-tree.jpg" alt="drawing" width="500px"/>


&emsp;&emsp;Regresia Decision Tree observ caracteristicile unui obiect i antreneaz un model n structura unui arbore pentru a prezice date n viitor si pentru a produce o ieire continu semnificativ. ## LassoLars

***Link catre documentatia oficiala***: [https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LassoLars.html](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LassoLars.html)

&emsp;&emsp;n statistici, regresia cu unghi minim (LARS) este un algoritm pentru adaptarea modelelor de regresie liniar la date de dimensiuni ridicate, dezvoltat de Bradley Efron, Trevor Hastie, Iain Johnstone i Robert Tibshirani.

&emsp;&emsp;S presupunem c ne ateptm ca o variabil de rspuns s fie determinat de o combinaie liniar a unui subset de covariabile poteniale. Apoi, algoritmul LARS ofer un mijloc de a produce o estimare a variabilelor de inclus, precum i a coeficienilor acestora.

&emsp;&emsp;n loc s dea un rezultat vector, soluia LARS const dintr-o curb care denot soluia pentru fiecare valoare a normei L1 a vectorului parametru. Algoritmul este similar cu regresia pas cu pas nainte, dar n loc s includ variabile la fiecare pas, parametrii estimate sunt crescui ntr-o direcie echiangular cu corelaiile fiecruia cu reziduul.

&emsp;&emsp;O explicaie simpl a regresiei Lasso i Least Angle

&emsp;&emsp;Dai un set de msurtori de intrare x1, x2 ... xp i o msurare a rezultatului y, lazo se potrivete unui model liniar

&emsp;&emsp;yhat = b0 + b1 * x1 + b2 * x2 + ... bp * xp

&emsp;&emsp;Criteriul pe care l folosete este:

&emsp;&emsp;Minimizai suma ((y-yhat) ^ 2) sub rezerva sumei [valoare absolut (bj)] <= s

&emsp;&emsp;Prima sum este preluat de observaiile (cazurile) din setul de date. S legat este un parametru de reglare. Cnd s este suficient de mare, constrngerea nu are efect i soluia este doar regresia obinuit multipl liniar obinuit a lui y pe x1, x2, ... xp.

&emsp;&emsp;Cu toate acestea, atunci cnd pentru valori mai mici de s (s> = 0) soluiile sunt versiuni micorate ale estimrilor celor mai mici ptrate. Adesea, unii dintre coeficienii bj sunt zero. Alegerea s este ca alegerea numrului de predictori de utilizat ntr-un model de regresie, iar validarea ncruciat este un instrument bun pentru estimarea celei mai bune valori pentru s.


* Calculul soluiilor Lasso

&emsp;&emsp;Calculul soluiilor lazo este o problem de programare ptratic i poate fi abordat prin algoritmi standard de analiz numeric. Dar procedura de regresie cu unghi minim este o abordare mai bun. Acest algoritm exploateaz structura special a problemei lazo i ofer o modalitate eficient de a calcula soluiile simultan pentru toate valorile s.

&emsp;&emsp;Regresia cu unghi minim este ca o versiune mai democratic a regresiei n trepte. Amintii-v cum funcioneaz regresia n trepte:

* Algoritmul de regresie pas cu pas nainte:

&emsp;&emsp;ncepei cu toi coeficienii bj egali cu zero.
&emsp;&emsp;Gsii predictorul xj cel mai corelat cu y i adugai-l n model. Luai reziduuri r = y-yhat.
&emsp;&emsp;Continuai, n fiecare etap adugnd la model predictorul cel mai corelat cu r.
&emsp;&emsp;Pn la: toi predictorii sunt n model
&emsp;&emsp;Cea mai mic procedur de regresie a unghiului urmeaz aceeai schem general, dar nu adaug un predictor pe deplin n model. Coeficientul acelui predictor este crescut numai pn cnd acel predictor nu mai este cel mai corelat cu r rezidual. Apoi, un alt predictor concurent este invitat s se alture clubului.


* Algoritmul de regresie cu unghi minim:

&emsp;&emsp;ncepei cu toi coeficienii bj egali cu zero.
&emsp;&emsp;Gsii predictorul xj cel mai corelat cu y
&emsp;&emsp;Cretei coeficientul bj n direcia semnului corelaiei sale cu y. Luai reziduuri r = y-yhat pe parcurs. Oprii-v cnd un alt predictor xk are la fel de mult corelaie cu r ca xj.
&emsp;&emsp;Cretei (bj, bk) n direcia articulaiei lor minime ptrate, pn cnd un alt predictor xm are la fel de mult corelaie cu r rezidual.
&emsp;&emsp;Continuai pn: toi predictorii sunt n model
&emsp;&emsp;n mod surprinztor, se poate demonstra c, cu o modificare, aceast procedur ofer ntreaga cale a soluiilor lazo, deoarece s variaz de la 0 la infinit. Modificarea necesar este: dac un coeficient diferit de zero atinge zero, eliminai-l din setul activ de predictori i recomputai direcia articulaiei.

Sen, care au publicat articole despre aceast metod n 1950 i respectiv n 1968  i dup Maurice Kendall datorit relaiei sale cu coeficientul de corelaie a rangului tau Kendall.

&emsp;&emsp;Acest estimator poate fi calculat eficient i este insensibil la valori anormale. Poate fi semnificativ mai precis dect regresia liniar simpl non-robust (cele mai mici ptrate) pentru datele nclinate i heteroskedastice i concureaz bine mpotriva celor mai mici ptrate chiar i pentru datele distribuite n mod normal n termeni de putere statistic. A fost numit cea mai popular tehnic neparametric pentru estimarea unei tendine liniare.

&emsp;&emsp;Panta median a unui set de n puncte de eantionare poate fi calculat exact calculnd toate O (n2) drepte prin perechi de puncte i apoi aplicnd un algoritm liniar de gsire a medianei timpului. Alternativ, poate fi estimat prin eantionarea perechilor de puncte. Aceast problem este echivalent, n cadrul dualitii proiective, cu problema gsirii punctului de trecere ntr-un aranjament de linii care are coordonata x median ntre toate aceste puncte de trecere.

&emsp;&emsp;Problema efecturii seleciei pantei exact, dar mai eficient dect algoritmul timpului ptratic al forei brute a fost studiat pe larg n geometria de calcul. Sunt cunoscute mai multe metode diferite pentru calcularea estimatorului Theil  Sen exact n timpul O (n log n), fie deterministic , fie folosind algoritmi randomizai. Estimatorul mediu repetat al lui Siegel poate fi, de asemenea, construit n acelai timp legat. n modelele de calcul n care coordonatele de intrare sunt numere ntregi i n care operaiunile bit-bit pe numere ntregi necesit timp constant, estimatorul Theil-Sen poate fi construit i mai rapid, n timp aleatorizat ateptat {\ displaystyle O (n {\ sqrt {\ log n}})} O (n {\ sqrt {\ log n}}).

&emsp;&emsp;Un estimator pentru panta cu rang aproximativ mediu, avnd acelai punct de defalcare ca i estimatorul Theil-Sen, poate fi meninut n modelul fluxului de date (n care punctele de eantionare sunt procesate unul cte unul de un algoritm care nu are suficient persisten stocare pentru a reprezenta ntregul set de date) folosind un algoritm bazat pe reele .
